# Data Cleaning

Pada langkah ini, kita akan mengisi data yang hilang dengan nilai lain

```python
# Melihat jumlah data yang hilang
df.isnull().sum()
```

```python
# Kita akan menghapus baris yang memiliki nilai hilang di kolom 'director'
df = df.dropna(subset=['director'], inplace=False)
print(df.info())
```

```python
# Kita akan menghapus baris yang memiliki nilai hilang di kolom 'country'
df = df.dropna(subset=['country'], inplace=False)
print(df.info())
```

```python
# Kita akan menghapus baris yang memiliki nilai hilang di kolom 'rating'
df = df.dropna(subset=['rating'], inplace=False)
print(df.info())
```

```python
# Melihat jumlah data yang hilang setelah dibersihkan
df.isnull().sum()
```

```python
# Membaca data
df.head()
```

((OPTIONAL)) Selain menghapus baris dengan nilai yang hilang, solusi alternatifnya adalah dengan mengisi nilai-nilai tersebut secara manual menggunakan `df.fillna(<value>)`.

```python
# upload ulang data awal dengan nilai yg hilang
df = pd.read_csv('/content/drive/My Drive/netflix_titles.csv')
# menampilkan baris dengan nilai yang hilang
df[df.isna().any(axis=1)]
```

```python
# Melihat jumlah data yang hilang setelah dibersihkan
df.isnull().sum()
```

```python
# mengisi nilai hilang di kolom country dengan nilai modus
df['country'] = df['country'].fillna(df['country'].mode()[0])
df[df.isna().any(axis=1)]
```

```python
# Melihat jumlah data yang hilang setelah dibersihkan
df.isnull().sum()
```

```python
# mengisi nilai hilang di kolom director dengan nilai modus
df['director'] = df['director'].fillna(df['director'].mode()[0])
df[df.isna().any(axis=1)]
```

```python
# Melihat jumlah data yang hilang setelah dibersihkan
df.isnull().sum()
```

```python
# melihat distribusi data berbandingkan dengan rating
df.groupby('rating').count()
```

```python
# mengisi nilai hilang di kolom rating dengan 'TV-MA'  -- lebih sering muncul sesuai hasil di atas
df['rating'] = df['rating'].fillna("TV-MA")
df[df.isna().any(axis=1)]
```

```python
# Melihat jumlah data yang hilang setelah dibersihkan
df.isnull().sum()
```

```python
# Melihat jumlah data duplikat
df.duplicated().sum()
```
